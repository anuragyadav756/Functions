###Q.6. Explain the concept of generators in Python and how they are defined.
- Generators in Python: Lazy Iteration
Generators in Python are a special kind of function that allows you to iterate over a potentially large sequence of values without storing the entire sequence in memory at once. They produce values on demand, making them memory-efficient, especially when dealing with very large datasets or infinite sequences.

Key Concepts:

- Lazy Evaluation: Generators employ lazy evaluation. They only compute and yield the next value in the sequence when it's specifically requested. This contrasts with regular functions that create and return an entire list or other iterable at once.
- yield Keyword: The core of a generator function is the yield keyword. When a yield statement is encountered, the function's execution is paused, and the value specified in the yield statement is returned to the caller. The function's state (including local variables and the point of execution) is saved, so that when the generator is asked for the next value, it resumes right after the yield statement.
- Iterable: Generators are iterators. This means you can use them in for loops, with the next() function, and in other contexts where an iterable is expected.
- Memory Efficiency: Because generators produce one item at a time, they use a fixed amount of memory, regardless of the size of the sequence they can potentially generate.

HOW GENERATORS ARE DEFINED:
- You define a generator function just like a regular Python function, but instead of using the return keyword to send back a final result, you use the yield keyword to produce a series of values.



###Q.7 What are the advantages of using generators over regular functions?
- Generators in Python offer several significant advantages over regular functions when dealing with sequences of data:

1. Memory Efficiency (Lazy Evaluation):

Regular Functions: When a regular function needs to return a sequence of values (e.g., a list), it typically generates the entire sequence in memory before returning it. For very large sequences, this can lead to significant memory consumption and potentially cause your program to crash or run very slowly due to memory limitations.
Generators: Generators employ lazy evaluation. They only produce the next value in the sequence when it's explicitly asked for (e.g., in a for loop or using next()). This means that only one item is held in memory at a time. For extremely large or even infinite sequences, this makes generators the only feasible option as they have a fixed memory footprint.
2. Improved Performance (Especially for Large Sequences):

Regular Functions: The overhead of creating and storing the entire sequence in memory can take a considerable amount of time, especially for large datasets.
Generators: Generators save computation time by only calculating values when they are needed. If you only need the first few items of a potentially long sequence, a generator will only compute those, whereas a regular function would compute the entire sequence unnecessarily.
3. Suitable for Infinite Sequences:

Regular Functions: A regular function trying to generate an infinite sequence would run indefinitely and likely exhaust memory.
Generators: Generators can easily represent infinite sequences because they only produce values on demand. You can iterate over a portion of an infinite sequence generated by a generator without any issues.
4. Easier to Implement Custom Iterators:

Regular Functions (for iterators): To create a custom iterator using regular functions, you would typically need to define a class with __iter__ and __next__ methods, manually manage the state of the iteration, and handle the StopIteration exception. This can be verbose and error-prone.
Generators: Generator functions provide a much more concise and natural way to implement custom iterators. The yield keyword automatically handles the state saving and the raising of StopIteration when the generator is exhausted. This makes the code cleaner and easier to understand.
5. Pipeline Processing:

Generators: Generators can be easily chained together to form data processing pipelines. The output of one generator can be the input of another, allowing for efficient and modular processing of data without creating intermediate lists in memory. This is a powerful paradigm for data manipulation.

###Q.9. Explain the purpose and usage of the `map()` function in Python.
- The map() function in Python is a built-in function that allows you to apply a given function to all items in an iterable (like a list, tuple, or string) and returns an iterator that yields the results.

Purpose:

The primary purpose of map() is to streamline the process of performing the same operation on each element of an iterable without the need for explicit loops (like for loops). It promotes a more functional programming style, making code more concise and often more readable for certain types of transformations.

Usage:

The map() function takes two main arguments:

- function: This is the function you want to apply to each item in the iterable. This function can be a built-in function, a user-defined function, or even a lambda function. It should typically take one argument (the current item from the iterable).

- iterable: This is the sequence (list, tuple, string, etc.) whose elements you want to process using the provided function. You can provide one or more iterables to map(). If you provide multiple iterables, the function must accept as many arguments as there are iterables, and map() will stop when the shortest iterable is exhausted.

###Q.10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The map(), reduce(), and filter() functions in Python are built-in higher-order functions that operate on iterables. They each serve a distinct purpose in transforming and processing data:

1. map():

- Purpose: Applies a given function to each item in an iterable and returns an iterator of the results.
- Operation: Transforms each element of the input iterable based on the provided function.
- Return Value: Returns a map object (an iterator). You typically need to convert it to a list, tuple, or other sequence type to see the results.
- Number of Inputs: Takes a function and one or more iterables. If multiple iterables are provided, the function must accept that many arguments, and map() stops when the shortest iterable is exhausted.
- Analogy: Think of it as applying a stencil (the function) to each page (element) of a book (iterable) to create a new book of modified pages.

2. filter():

- Purpose: Filters elements from an iterable based on a given function (the predicate). It returns an iterator containing only the elements for which the function returns True.
- Operation: Tests each element of the input iterable with the provided function. If the function returns True, the element is included in the result; otherwise, it's excluded.
- Return Value: Returns a filter object (an iterator). You usually need to convert it to a list, tuple, or other sequence type.
Number of Inputs: Takes a function (that should return a boolean value) and one iterable.
- Analogy: Think of it as using a sieve (the function) to separate the desired grains (elements that satisfy the condition) from the rest.
3. reduce():

- Purpose: Applies a given function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value.
- Operation: Takes the first two items in the iterable and applies the function to them. Then, it applies the function to the result and the next item, and so on, until only a single result remains.
- Return Value: Returns a single value (not an iterator).
- Number of Inputs: Takes a function (that accepts two arguments) and one iterable. It can also take an optional third argument, the initializer, which is placed before the items of the iterable in the calculation and serves as a default starting value.
- Analogy: Think of it as folding a piece of paper multiple times (applying the function), where each fold combines the current state with the next part of the paper until you have a single, final fold.

```python
###4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print(primes)
```

```python
###5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.num_terms:
            fib_number = self.current
            self.current, self.next = self.next, self.current + self.next
            self.count += 1
            return fib_number
        else:
            raise StopIteration

# Example usage:
fib_iter = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fib_iter:
    print(num)
```

```python
###6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

#Example
for power in powers_of_two(5):
    print(power)
```

```python
###7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        print(f"Error: File not found at path: {file_path}")

# Example usage:
# First, let's create a sample file named 'my_file.txt'
with open('my_file.txt', 'w') as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("And this is the third line.\n")

# Now, let's use the generator function to read it
file_reader = read_file_lines('my_file.txt')

print("Reading file line by line:")
for line in file_reader:
    print(f"Line read: '{line.strip()}'")  # .strip() removes leading/trailing whitespace

# Example of handling FileNotFoundError
file_reader_error = read_file_lines('non_existent_file.txt')
for line in file_reader_error:
    print(line) # This loop will not execute as the generator didn't yield anything
```

```python
###10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"  # Vowels to be removed (both lowercase and uppercase)
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Using filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Output the result
print(filtered_string)
```

```python
###11.11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

#Order Number      Book Title and Author                        Quantity              Price  Per Item
#34587             Learning Python, Mark Lutz                       4                    40.95
#98762             Programming Python, Mark Lutz                    5                    56.80
#77226             Head First in Python, Paul Barry                 3                    32.95
#88112             Einfhrung in Python3, Bernd Klein               3                    24.99

#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
#order is smaller than 100,00 .

#Write a Python program using lambda and map.

orders_data = [
    "#34587   Learning Python, Mark Lutz               4       40.95",
    "#98762   Programming Python, Mark Lutz            5       56.80",
    "#77226   Head First in Python, Paul Barry         3       32.95",
    "#88112   Einfhrung in Python3, Bernd Klein       3       24.99",
]

def process_order(order_string):
    """Processes a single order string to extract order number and calculate adjusted total."""
    parts = order_string.split()
    order_number = parts[0].lstrip('#')
    quantity = int(parts[-2])
    price_per_item = float(parts[-1])
    total_price = quantity * price_per_item
    if total_price < 100.00:
        total_price += 10.00
    return (order_number, round(total_price, 2))

# Process the orders using map and a lambda function
processed_orders = list(map(lambda order: process_order(order), orders_data))

print(processed_orders)
```
